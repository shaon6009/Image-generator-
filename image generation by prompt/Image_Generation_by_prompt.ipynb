```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
from PIL import Image


modelid = "CompVis/stable-diffusion-v1-4"
device = "cuda" if torch.cuda.is_available() else "cpu"

auth_token = "your_auth_token_here"
pipe = StableDiffusionPipeline.from_pretrained(modelid, revision="fp16", use_auth_token=auth_token)
pipe.to(device)

def generate_image(prompt_text, height=768, width=768, guidance_scale=10.0):
    with torch.autocast(device):
        image = pipe(prompt_text, height=height, width=width, guidance_scale=guidance_scale).images[0]

    return image

prompt = input("Enter your prompt for image generation (be specific and detailed): ")

if prompt:
    generated_image = generate_image(prompt)
    plt.imshow(generated_image)
    plt.axis("off")
    plt.show()
else:
    print("Please enter a valid prompt!")
```

বাড়ানো (Increase): যদি guidance_scale বাড়ানো হয় (যেমন 9.0 থেকে 12.0 করা হয়), তাহলে ছবিটি আপনার প্রদত্ত প্রম্পটের সাথে আরও বেশি মেলে। তবে, কখনও কখনও এটি ছবির মানকে খারাপ করতে পারে বা কিছু অস্বাভাবিক বৈশিষ্ট্য তৈরি করতে পারে।

কমানো (Decrease): যদি guidance_scale কমানো হয় (যেমন 7.0 করা হয়), তাহলে ছবিটি আরও সৃজনশীল হতে পারে এবং কিছু নতুন বৈশিষ্ট্য যুক্ত হতে পারে, কিন্তু এটি প্রম্পটের সাথে কম সম্পর্কিত হতে পারে।
